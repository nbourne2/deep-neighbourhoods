# LST Signal-to-Noise Analysis

LST maps from the builder are constructed by combining a stack of images. There are no explicit uncertainty maps associated with them, but a measure of uncertainty on the Mean LST can be obtained simply from the standard error of data being averaged in each pixel.

This notebook is intended to explore the signal-to-noise ratio of the $LST$ and $r LST$ maps as estimated from the standard error, ranging from the raw resolution up to LSOA aggregation level.

Think about it: SNR is not quite right: the temperature can be 0, or negative (true for both $LST$ and $r LST$) which would obviously lead to low SNR.  The "signal" in SNR is really $\Delta LST$ ($\equiv \Delta rLST$), so we need to quantify the SNR of the contrast, not of the LST itself.

So there's no point looking at raw $LST$. Best to look at $r LST$. And there's limited value in an SNR map, because extremes of temperature will have high (positive or negative) SNR when in fact they could still have large uncertainties.  Probably what we're interested in is the pixel distribution of the SNR map, to ensure that the range and typical variations in the mean LST are large in comparison to "uncertainty" in the pixel values (where uncertainty here means standard error).

**Inputs:**
* LST/rLST mean, variance, and count images from LSOA_LST_builder
* LSOA polygons to aggregate within

**Process:**
* Produce standard error map, masking any pixels where count <=1
* Smooth images with 2x2 and 3x3 boxcar kernels and produce std error maps
* Make SNR maps 
* Produce standard error map at LSOA level (masking in similar way to LSOA_LST_builder)

**Outputs:**
* Somehow visualise these maps on a meaningful scale
* Show standard error map alongside mean $r LST$ map, with a good readable color scale
* Show map of $r LST$/error, alongside map of $r LST$, to understand how contrast in the map compares to the size of the errors.
* Somehow quantify the quality of LST measurements at LSOA level
