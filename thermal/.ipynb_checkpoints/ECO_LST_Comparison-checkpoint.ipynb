# Compare LST before/after ECO installations

Use EON dataset to locate particular ECO installations in a given year. Display LST maps of appropriate areas and do some analysis to determine whether any measurable effect of ECO on LST in comparison with other areas.

**Inputs**
* EON ECO installations data table
* LST maps for appropriate years from LSOA_LST_Builder

**Considerations**
* Define small AOI(s) located where lots of installations were done
* Be choosy about which installations to look at - initially just loft insulation
* When looking at control sample of LSOA's without ECO installations, try to choose similar geographical areas (look at land use map for example and go for areas nearby and with similar level of urbanity)



```python
"""Initialise"""
import os
import glob
import pprint as pp

import pandas as pd
import geopandas as gpd
import numpy as np
from numpy import ma as ma
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import geoplot as gpl
```

```python
"""Define dates of interest"""

min_date = '2014-02-28'
max_date = '2016-11-01'

before_label = '2013-2014'
after_label = '2016-2018'
place_label = 'derbyshire'

use_medsub_lst = True 
product_label = ('rLST' if use_medsub_lst else 'LST')
```

```python
# Where to find the files
rootdir = '/Users/nathan.bourne/data/thermcert/'

lsoa_before_file = rootdir+'output_LSOA_LST/{0}_{1}/lsoa_{0}_{2}_LC.geojson'.format(
                    place_label,before_label,product_label)
lsoa_after_file = rootdir+'output_LSOA_LST/{0}_{1}/lsoa_{0}_{2}_LC.geojson'.format(
                    place_label,after_label,product_label)

eco_file = rootdir+'eon/data/20180419_derbyshire_sheet1.csv'
pc_cent_file = rootdir+'uk_data/postcode_centroids/ONS_Postcode_Directory_Latest_Centroids.csv'
pc_dist_file = rootdir+'uk_data/astrosat_data/uk-stats/data/raw/geometries/postcode_district/Districts.shp'
```

```python
"""Read ECO data"""
eco_data = pd.read_csv(eco_file)
```

```python
"""Read postcode locations"""
pc_cent = pd.read_csv(pc_cent_file)

pcx = pd.to_numeric(pc_cent['X'],errors='coerce')
pcy = pd.to_numeric(pc_cent['Y'],errors='coerce')

"""Convert the Postcode Dataframe into a Geodataframe by adding geometry"""
from shapely.geometry import Point

pc_geom = list(zip(pcx,pcy))
pc_geom = [Point(xy) for xy in pc_geom]
pc_cent_gdf = gpd.GeoDataFrame(pc_cent,geometry=pc_geom)

# Currently no CRS
print(pc_cent_gdf.crs)
```

```python
"""Attach CRS to postcodes geodata"""
pc_dist = gpd.read_file(pc_dist_file)
print(pc_dist.crs)

pc_cent_gdf.crs = pc_dist.crs
```

```python
"""Read LSOA geodata"""

lsoa_lst_after = gpd.read_file(lsoa_after_file) 
lsoa_lst_before = gpd.read_file(lsoa_before_file) 

"""Check CRS"""
print(lsoa_lst_after.crs)
print(pc_cent_gdf.crs)
```

```python
"""Match postcodes CRS to LSOA CRS"""
pc_cent_gdf = pc_cent_gdf.to_crs(lsoa_lst_after.crs)
print(pc_cent_gdf.crs)
```

```python
"""Reformat postcodes to ensure common format"""
#eco_data = eco_data.replace('APPPOSTCODE', eco_data['APPPOSTCODE'].str.replace(' ',''))
#pc_cent_gdf = pc_cent_gdf.replace('pcd', pc_cent_gdf['pcd'].str.upper().str.replace(' ',''))
eco_data['APPPOSTCODE'] = eco_data['APPPOSTCODE'].str.upper().str.replace(' ','')
pc_cent_gdf['pcd'] = pc_cent_gdf['pcd'].str.upper().str.replace(' ','')
```

```python
"""Join ECO dataset to postcode geodata so have location info"""
eco_geodata = pd.merge(pc_cent_gdf,eco_data,left_on='pcd',right_on='APPPOSTCODE',how='right')
```

```python
"""Obtain dates of ECO installations"""
install_dates = pd.to_datetime(eco_geodata['INSTALLDATE'])
```

```python
"""Filter installations"""

from shapely.geometry.point import Point

"""Valid geometry data"""
valid_points = np.array([type(geom) == Point for geom in eco_geodata['geometry']])
print('Valid geometry: {}/{}'.format(np.sum(valid_points),np.size(valid_points)))

"""Valid dates and within the range of interest"""
valid_dates = (pd.to_numeric(install_dates).values > 0)
valid_dates = ((install_dates > pd.to_datetime(min_date))
                          & (install_dates < pd.to_datetime(max_date)))
print('Valid dates: {}/{}'.format(np.sum(valid_dates),len(install_dates)))

"""Filter by installation type"""
valid_installs = ((eco_geodata['DESCRIPTION'].str.lower().str.contains('loft insulation'))
                  | (eco_geodata['DESCRIPTION'].str.lower().str.contains('flat roof insulation'))
#                   | (eco_geodata['DESCRIPTION'].str.lower().str.contains('cavity wall'))
#                   | (eco_geodata['DESCRIPTION'].str.lower().str.contains('external solid wall'))
                 )

print('Valid installations: {}/{}'.format(np.sum(valid_installs),len(valid_installs)))

"""Final filter"""
valid_records, = np.where(valid_dates & valid_installs & valid_points)
print('Combined filter: {}/{}'.format(len(valid_records),len(install_dates)))
```

```python
"""Trends"""

Nmax = 21
mean_LST = np.zeros(Nmax)+np.nan
stde_LST = np.zeros(Nmax)+np.nan
p50_LST = np.zeros(Nmax)+np.nan
p25_LST = np.zeros(Nmax)+np.nan
p75_LST = np.zeros(Nmax)+np.nan
N_data = np.zeros(Nmax).astype(np.int)
for ii in range(Nmax):
    ind, = np.where(lsoa_geodata_grouped['ECO_count'] == ii)
    N_data[ii] = np.sum(np.isfinite(lsoa_geodata_grouped['LST_mean_diff_medsub'][ind]))
    if N_data[ii]>2:
        mean_LST[ii] = np.nanmean(lsoa_geodata_grouped['LST_mean_diff_medsub'][ind])
        stde_LST[ii] = (np.nanstd(lsoa_geodata_grouped['LST_mean_diff_medsub'][ind]) /
                        np.sqrt(N_data[ii])
                       )
    if N_data[ii]>2:
        (p25_LST[ii],p50_LST[ii],p75_LST[ii]) = np.nanpercentile(lsoa_geodata_grouped['LST_mean_diff_medsub'][ind],
                                                 [25,50,75])
    
fig,axes = plt.subplots(figsize=(15,7))
axes.errorbar(np.arange(Nmax),mean_LST,stde_LST,linestyle='None',marker='o',color='b')


axes.errorbar(np.arange(Nmax)+0.25,p50_LST,[p50_LST-p25_LST,p75_LST-p50_LST],
               color='0.5',marker='s',linestyle='None')
for ii in np.arange(Nmax):
    if N_data[ii]>2:
        axes.text(ii,-0.35,N_data[ii])
axes.text(0,-0.32,'Number of LSOAs:')
# Attempt a fit
# scipy.optimize.curve_fit(f, xdata, ydata, p0=None, sigma=None, absolute_sigma=False, check_finite=True, bounds=(-inf, inf), method=None, jac=None, **kwargs)
from scipy.optimize import curve_fit

def linfun(x,p0,p1):
    return x*p1 + p0

xfit = np.arange(Nmax)
notnan = (np.isfinite(mean_LST) & np.isfinite(stde_LST) )
par,pcov = curve_fit(linfun, xfit[notnan], mean_LST[notnan], sigma=stde_LST[notnan],
                    absolute_sigma=True)
print,par
yfit = linfun(xfit,*par)
axes.plot(xfit[notnan],yfit[notnan],color='b',linestyle='--')

fitstr = 'gradient = {:5.3f} $\pm$ {:5.3f}; intercept = {:5.3f} $\pm$ {:5.3f}'.format(
            par[1],pcov[1,1]**0.5,par[0],pcov[0,0]**0.5)
axes.text(1,0.2,fitstr,color='b')

axes.set_ylabel('$\Delta$ r LST')
axes.set_xlabel('Number of valid ECO installations in LSOA')
```

**So there is a measurable correlation between number of installations and $\Delta$rLST, AND there is a statistically significant difference between those areas with installations and those with none**
